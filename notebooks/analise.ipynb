```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
print("Ambiente configurado com sucesso 🚀")
```

```python
df_vendas = pd.read_csv("../data/Dados_AnaliseXTFAS.csv")

df_vendas.head()
```

```python
df_vendas.info()
```

```python
df_vendas.describe()
```

```python
df_vendas["Data"] = pd.to_datetime(df_vendas["Data"])

df_vendas.info()
```

```python
faturamento_total = df_vendas["Total"].sum()
print(f"Faturamento Total: R$ {faturamento_total:,.2f}")
```

```python
produto_mais_vendido = df_vendas.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)

produto_mais_vendido
```

```python
receita_por_produto = df_vendas.groupby("Produto")["Total"].sum().sort_values(ascending=False)

receita_por_produto
```

```python
receita_por_produto.plot(kind="bar")

plt.title("Receita por Produto")
plt.xlabel("Produto")
plt.ylabel("Receita Total")
plt.xticks(rotation=45)

plt.show()
```

📊 Análise de Receita por Produto

Observa-se que o Produto B é o maior gerador de receita, totalizando R$ 3.750.

Apesar de o Produto A ter sido o menos vendido em quantidade, ele aparece como o segundo maior gerador de receita, indicando que possui preço unitário mais elevado.

Esse comportamento pode indicar uma estratégia de produto premium ou maior margem de lucro.

```python
vendas_por_data = df_vendas.groupby("Data")["Total"].sum()

vendas_por_data.head()
```

```python
import matplotlib.pyplot as plt

vendas_por_data.plot()

plt.title("Evolução da Receita ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Receita")
plt.xticks(rotation=45)

plt.show()
```

📊 Padrão Cíclico de Receita

A análise temporal demonstra um comportamento cíclico consistente ao longo do período observado.

Observa-se alternância frequente entre picos e quedas na receita diária, sugerindo um padrão recorrente de consumo. Esse comportamento pode estar associado a:

Sazonalidade de curto prazo

Padrão fixo de reposição de estoque

Estratégias comerciais recorrentes

A identificação desse padrão permite melhor planejamento financeiro e definição de estratégias comerciais mais eficientes.

```python
ticket_medio = df_vendas["Total"].mean()
print(f"Ticket médio: R$ {ticket_medio:.2f}")
```

## 📌 Conclusão

A análise exploratória permitiu identificar padrões relevantes no comportamento das vendas.

Os principais insights observados foram:

- O faturamento total no período analisado foi de **R$ 8.750**.
- O **Produto B** foi responsável pelo maior volume de vendas e maior geração de receita.
- A análise temporal revelou um comportamento cíclico na receita ao longo do período.
